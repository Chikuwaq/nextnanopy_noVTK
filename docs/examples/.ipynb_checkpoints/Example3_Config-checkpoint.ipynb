# Default configuration

nextnano configuration is stored in .nncofig file located at your home directory (Usually: C:\Users\Your_User)

When you import nextnanopy, it will generate automatically the configuration file with few default parameters.

This configuration file is useful if you want to load any input file and execute them without passing the compulsory arguments like executable, license, database, etc.

```python
nn.config
```

## Set your own configuration

Please, use absolute paths to avoid any possible problem.

```python
nn.config.set('nextnano++','exe',r'C:\Program Files\nextnano\2020_07_31\nextnano++\bin 64bit\nextnano++_Intel_64bit.exe')
nn.config.set('nextnano++','license',r'C:\Users\jun-liang.wang\Documents\nextnano\License\License_nnp.lic')
nn.config.set('nextnano++','database',r'C:\Program Files\nextnano\2020_07_31\nextnano++\Syntax\database_nnp.in')
nn.config.set('nextnano++','outputdirectory',r'E:\junliang.wang\nextnano\Output')

nn.config.set('nextnano3','exe',r'C:\Program Files\nextnano\2020_07_31\nextnano3\Intel 64bit\nextnano3_Intel_64bit.exe')
nn.config.set('nextnano3','license',r'C:\Users\jun-liang.wang\Documents\nextnano\License\License_nnp.lic')
nn.config.set('nextnano3','database',r'C:\Program Files\nextnano\2020_07_31\nextnano3\Syntax\database_nn3.in')
nn.config.set('nextnano3','outputdirectory',r'E:\junliang.wang\nextnano\Output')

nn.config.save() #save permanently
```

```python
nn.config
```

# Execute input with default config

By loading an input file with InputFile class, it will automatically detect the type (nextnano++ or nextnano3) and it loads the corresponding configuration parameters ready for execution.

```python
file = os.path.join(folder_nnp,'example.in')
fi = nn.InputFile(file) # it will load automatically the config saved at home directory
```

```python
print(f'Config file location: {fi.config.fullpath}')
```

```python
fi.config
```

## Detected config parameters based on input file type

```python
print(f'File type: {fi.type}')
print(f'execute arguments: {fi.default_command_args}')
```

## Execute with default configuration

```python
fi.execute() # it works without any arguments
```

## Execute with one or more temporary parameters

```python
fi.execute(threads=4,outputdirectory=r'C:\Users\jun-liang.wang\Downloads')
```
